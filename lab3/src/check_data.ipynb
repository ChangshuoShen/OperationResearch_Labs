```python
import numpy as np

file_path = '../other/a9a.txt'
num_features = 123


labels = []
features = []

# Process each line in the file
with open(file_path, 'r') as file:
    for line in file:
        # Split line into individual entries
        parts = line.strip().split()
        # First element is the label
        labels.append(int(parts[0]))
        # Remaining parts are feature:value pairs
        feature_row = np.zeros(num_features)
        for item in parts[1:]:
            index, value = item.split(':')
            feature_row[int(index) - 1] = float(value)  # Convert index to 0-based
        features.append(feature_row)
        
# Convert to numpy arrays
X = np.array(features)
y = np.array(labels)
    
X.shape, y.shape, np.sum(y)
```

```python
import pandas as pd
df = pd.DataFrame(X[:10], columns=[f"Feature {i+1}" for i in range(num_features)])
df["Label"] = y[:10]

df.describe()
```
